For a first order **kinematic** Model: 

$\quad \dot{x} (t) = u(t)$

$\quad x(t) \in \mathbb{R}  : State $

$\quad u(t) \in \mathbb{R}  : Control \, Input $


### Problem: 
We meassure $x(t)$
Design a control law  or controller to find $u(t)$ to make $x(t) \to x_d (t)\; as\; t\to \infty$

#### Solution:
What we really want is $e(t) := x_d - x(t) \longrightarrow 0 \quad as \; t\to \infty$

Look at the **Error Dynamics**

$\quad \frac{de(t)}{dt} = \dot{e}(t) = 0 - \dot{x}(t) = -u(t)$

Pick a Controller Design 
$\quad u(t) = K (x_d - x(t)) = k e(t)$

$\quad \dot{e}(t) = -k e(t) :[Eq.1] \longrightarrow $  This is **Proportional Control** 

How to pick K? 

Find K so that $e(t) \to 0 \; as t \to \infty$

#### How to solve [Eq.1]

One way is through the **Laplace Transform**

$ Let \; \mathcal{L}[e(t)] = E(s)  $

$ \mathcal{L}[\dot{e}(t)] = \mathcal{L}[-k e(t)] = -K \mathcal{L}[e(t)]  $

$ (s+k)E(s) = e(0) $

$ E(s) = \frac{e(0)}{s+k} $

$ \mathcal{L}^-1 [E(s)] = e(t) = \mathcal{L}^-1 [\frac{e(0)}{s+k}] = e^{-kt} e(0) $

### Second Order Systems

Now suppose our vehicle model is now represented by 

$ \quad \ddot{x} (t) = \frac{1}{m} u(t) : Where \begin{cases} u(t) \; is\; Force \\ m \; is \; mass \\ \ddot{x} \; is \; Acceleration \end{cases} $ 


Again, How to choose $u(t)$ so that $x \longrightarrow x_{d}$ as $t\to\infty$

Let's write our model in **State Variable** form: 

$Let: \\ \quad x_1 = x \; (Position) \\ \quad x_2 = \dot{x} \; (Velocity)$

Thus: $\\ \quad \dot{x_1}=x_2 \\ \quad \dot{x_2}=\frac{1}{m}u(t)$

We can write this in matrix format 

$ \begin{bmatrix} \dot{x_1} \\ \dot{x_2} \end{bmatrix} = \underbrace{\begin{bmatrix} 0&1 \\ 0&0 \end{bmatrix}}_A  \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} + \underbrace{\begin{bmatrix} 0 \\ \frac{1}{m} \end{bmatrix}}_{B}u(t)$

Let: 

$ \quad \underline{\dot{x}} = A \underline{x} + B u(t) \quad Where \; \underline{x} = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} $ 

This is called the **Linear State Space Model**

#### Aside on stability ... 

**Note** : $\underline{x} = 0 $ is called an **Equilibrium point of $(1)$ because if we start at $\underline{x} =0 $ then it follows that $\underline{\dot{x}} =0 $ and therefore our system $(1)$ never changes. 

**Definition**: $\underline{Stability} \\ \quad$ The equilibrium point $\underline{x}=0$ is : 

$(i)\; \underline{Stable}$ if for every $\varepsilon>0$ there is a $\delta = \delta(\varepsilon)>0$ such that: 
$$ \|\underline{x}(0)\| < \delta \Rightarrow \|x(t)\| < \varepsilon \quad \forall t>0$$

### Theorem : Linear Stability

The Equilibrium point $\,\underline{x}=0\,$ of $\,\underline{\dot{x}}=F \underline{x}\,$  is **asymptotically Stable** $$\mathcal{iff} \; the \; eigenvalues, \; \lambda_{i}, \; satisfy \; $\mathcal{Re}(\lambda _i)< 0 $$


In otherwords, F has eigenvalues in the Left-hand Complex Plane (LHCP)


$F = A+BK = \begin{bmatrix} 0&1 \\ 0&0 \end{bmatrix} + \begin{bmatrix} 0 \\ \frac{1}{m} \end{bmatrix} \begin{bmatrix} k_1 & k_2 \end{bmatrix}  = \begin{bmatrix} 0&1 \\ \frac{k_1}{m} & \frac{k_2}{m} \end{bmatrix}$

Find the eigenvalues: 

$\det{\mathcal{I}\lambda-F} = \det {\begin{bmatrix} \lambda & -1 \\ \frac{-k_1}{m} & \frac{\lambda - k_2}{m} \end{bmatrix} } = p(\lambda) \quad $ where $p$ is the **Characteristic Polynomial**.

$p(\lambda)=\lambda^2 - \frac{\lambda k_2}{m} - \frac{k_1}{m} =0 \quad \longleftarrow $ Roots are the eigenvalues of $F$.

Choose $k_1$ and $k_2$ so that $\lambda_i$ have $\mathcal{Re}(\lambda_{i})<0$ as per theorem.

**Take Away**: if F has eigenvalues that are all within the Lefthand Complex plane then the Second Order System $(1)$ is **Asymptotically Stable**.
